```python
# Set the style for better visualization
plt.style.use('dark_background')
sns.set_palette("husl")



# Function to plot distribution of numerical features
def plot_numerical_distributions(dataframe):
    # Select numerical columns
    numerical_features = dataframe.select_dtypes(include=['float64', 'int64']).columns
    
    # Calculate number of rows and columns for subplots
    n_features = len(numerical_features)
    n_rows = (n_features + 2) // 3  # 3 plots per row
    n_cols = min(3, n_features)
    
    # Create subplots
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 4*n_rows))
    fig.suptitle('Distribution of Numerical Features in Liver Dataset', fontsize=16, y=1.02)
    
    # Flatten axes array for easier iteration
    axes = axes.flatten()
    
    # Plot distribution for each numerical feature
    for idx, feature in enumerate(numerical_features):
        # Histogram with KDE
        sns.histplot(data=dataframe, x=feature, kde=True, ax=axes[idx])
        axes[idx].set_title(f'Distribution of {feature}')
        axes[idx].set_xlabel(feature)
        axes[idx].set_ylabel('Count')
        
    # Remove empty subplots if any
    for idx in range(len(numerical_features), len(axes)):
        fig.delaxes(axes[idx])
    
    plt.tight_layout()
    plt.show()

# Function to create box plots
def plot_numerical_boxplots(dataframe):
    # Select numerical columns
    numerical_features = dataframe.select_dtypes(include=['float64', 'int64']).columns
    
    # Create figure
    plt.figure(figsize=(8, 6))
    
    # Create box plot
    sns.boxplot(data=dataframe[numerical_features])
    plt.xticks(rotation=45)
    plt.title('Box Plots of Numerical Features')
    plt.tight_layout()
    plt.show()

# Function to create correlation heatmap
def plot_correlation_heatmap(dataframe):
    # Select numerical columns
    numerical_features = dataframe.select_dtypes(include=['float64', 'int64'])
    
    # Calculate correlation matrix
    correlation_matrix = numerical_features.corr()
    
    # Create figure
    plt.figure(figsize=(8, 6))
    
    # Create heatmap
    #https://www.google.com/search?q=mask%3Dnp.triu(np.ones_like(corr%2Cdtype%3Dbool))&rlz=1C1GCEA_enIN1093IN1093&oq=mask%3Dnp.triu(np.ones_like(corr%2Cdtype%3Dbool))&gs_lcrp=EgZjaHJvbWUyBggAEEUYOdIBBzYxNGowajeoAgiwAgE&sourceid=chrome&ie=UTF-8
    # usually this code needs to be copy pasted as it has no maths reasoning behind it
    corr=dataframe.corr(numeric_only=True)
    
    mask=np.triu(np.ones_like(corr,dtype=bool))
    f,ax=plt.subplots(figsize=(11,9))
    cmap=sns.diverging_palette(250,5,as_cmap=True)
    sns.heatmap(corr,mask=mask,cmap=cmap,vmax=3,center=0,square=True,linewidths=5,cbar_kws={"shrink":.5})#cbar_kws is a argument that sepcifies options fro the colorbar .in this case its shrink the size of it to 0.5 

    # sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Heatmap of Numerical Features')
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    try:
        # Plot distributions
        plot_numerical_distributions(liver_data)
        
        # Plot box plots
        plot_numerical_boxplots(liver_data)
        
        # Plot correlation heatmap
        plot_correlation_heatmap(liver_data)
        
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
y=liver_data["Selector"]
remove=["Selector","Age_Group"]
X=liver_data.drop(remove,axis=1)  #q-- y didnt "index" work here at axis= than 1 working ?...this is ans at cell {118}

X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train["Gender"].hist()
```

```python
y_train_norm=(y_train==1)
y_test_norm=(y_train==1)
```

```python
print("\nColumns with missing values:")
print(liver_data.columns[liver_data.isnull().any()].tolist())
```

```python
X_train.iloc[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
k = 2
knn = KNeighborsClassifier(n_neighbors=k)

knn.fit(X_train, y_train)


y_pred = knn.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)

```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred=cross_val_predict(knn,X_train,y_train_norm,cv=3)
y_train_pred
```

```python
from sklearn.metrics import confusion_matrix
# the metric module contains various functions to evaluate the performance of machine learning models
confusion_matrix(y_train_norm,y_train_pred)
```
